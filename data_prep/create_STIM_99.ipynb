In consultation with my supervisor, I decided to present the 17 stimuli which only produced 2 validated interactions to two additional experimental subjects on Prolific. The goal of this is to increase the number of validated interactions for these stimuli, so that more aggregate human attribution maps can be created. (These are produced by averaging at least three individual human attribution maps as described in the following section.) For this purpose, a new stimuli file, stimuli_99.pickle, was created alongside the original 3 sets of stimuli, and uploaded to the online interface and deployed on Prolific.
