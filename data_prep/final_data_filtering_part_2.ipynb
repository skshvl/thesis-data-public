# Determine how many of each type of data to sample

```python
sampling_amounts = pd.read_excel("sampling_goal_amounts.xlsx", engine="openpyxl", header = 1)
```

We can safely **remove existence from this list** because we simply want all approved existence pieces (there are exactly 33)

```python
sampling_amounts = sampling_amounts[sampling_amounts["linguistic_phenomena"]!="existence"]
sampling_amounts
```

# Start building sample

```python
samples = []

#add "existence" already
samples.append(df[(df["Approval"]=="Approve") & (df["linguistic_phenomena"]=="existence")])
```

```python
# df[(df['linguistic_phenomena']=="actions") & (df['Approval']=="Approve") & (df['performance_group']=="high_perf")]
```

```python
for index, row in sampling_amounts.iterrows():
    print("Selecting rows based on criteria:", dict(row))
    selected_rows = df[(df['linguistic_phenomena']==row['linguistic_phenomena']) & (df['Approval']==row['Approval']) & (df['performance_group']==row['performance_group'])]
    print(f"Found {len(selected_rows)} instances")

    if len(selected_rows) < row['sample']:
        print(f">>Not enough examples of selected criteria to sample {row['sample']} cases")

    samples.append(selected_rows.sample(row['sample'], random_state=42))
```

### Concatenate all sampled dataframes

```python
stimuli_df = pd.concat(samples)
stimuli_df.groupby(["linguistic_phenomena",
                    "performance_group",
                    "Approval"
                        ]   ).count().to_excel("actions_relations_sample_numbers.xlsx")
```

# Export all stimuli into one CSV

# Create 3 subgroups of stimuli (start here)

```python
stimuli_df = pd.read_csv("all_stimuli_final.csv")
stimuli_df.head(2)
```

### Use stratified sampling to balance performance_group and linguistic_phenomena

```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold

#  stratify by linguistic_phenomena to ensure each type has exactly 11 examples in each subset
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
stratified_dfs = []

# new column for stratification -- combines both features: linguistic phenomena and performance group
stimuli_df['stratify_col'] = stimuli_df['linguistic_phenomena'].astype(str) + "_" + stimuli_df['performance_group'].astype(str)

for train_index, test_index in skf.split(stimuli_df, stimuli_df['stratify_col']):
    stratified_dfs.append(stimuli_df.iloc[test_index])

# drop the temp column
stimuli_df.drop('stratify_col', axis=1, inplace=True)

# Assign each subset to a new DataFrame
stimuli_1_df = stratified_dfs[0].reset_index(drop=True)
stimuli_2_df = stratified_dfs[1].reset_index(drop=True)
stimuli_3_df = stratified_dfs[2].reset_index(drop=True)

# also drop the temp column from the final three dataframes
for stim_df in [stimuli_1_df, stimuli_2_df, stimuli_3_df]:
    stim_df.drop('stratify_col', axis =1, inplace = True)
```
