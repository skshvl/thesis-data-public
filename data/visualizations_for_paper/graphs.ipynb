```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Determine whether approve/reject of stimuli correlates with CLIP performance

Using EXCEL of 269 labeled stimuli

```python
stimuli_df = pd.read_excel("../../data_prep/269_stimuli_edited_labeled.xlsx")
stimuli_df.head(1)
```

Actions and relations

```python
sns.set_theme(style="darkgrid")

def plot_linguistic_phenomenon(df, phenomenon):
    """Create categorical scatter plot with Approval status on x axis and clip_pred_difference on y axis"""
    subset = df[df['linguistic_phenomena'] == phenomenon]

    #  scatter plot with 'Approval' on x-axis and 'clip_pred_difference' on y-axis
    g = sns.catplot(x='Approval', y='clip_pred_diff', data=subset,
                    kind='strip', dodge=True, edgecolor='none', alpha=0.5, s=15, height=6, aspect=1.5)

    #  mean and standard deviation
    means = subset.groupby('Approval')['clip_pred_diff'].mean()
    stds = subset.groupby('Approval')['clip_pred_diff'].std()
    ax = g.ax
    for i, (approval, mean) in enumerate(means.items()):
        ax.errorbar(approval, mean, yerr=stds[i], fmt='o', color='r', capsize=5)

    ax.set_title(f'Scatter Plot for {phenomenon}')
    ax.set_xlabel('Approval')
    ax.set_ylabel('Clip Prediction Difference')


plot_linguistic_phenomenon(stimuli_df, "actions")
plot_linguistic_phenomenon(stimuli_df, "relations")
```

```python
plot_linguistic_phenomenon(stimuli_df, "existence")
```
