```python
import pandas as pd
import os

models = ["LXMERT", "CLIP", "FLAVA", "SigLip"]

#put all models' VALSE output data into a single dataframe
dfs = [pd.read_pickle(f"../xai_maps/xai_output/{model_name}_all_of_valse_output_scores.pickle") for model_name in models]
for model_name, model_df in zip(models, dfs):
    model_df['model']=model_name
df = pd.concat(dfs)

df['filename'] = [os.path.basename(path) for path in df['img_path']]

df.head(1)
```

### Graph accuracy for each model

```python
import matplotlib.pyplot as plt
import seaborn as sns


def graph_performance(df, model_name="all models"):
    # set seaborn style
    sns.set_theme(style="dark")

    # filter for specific model or all models
    if model_name == "all models":
        model_name = "ALL MODELS (!)"  # update caption for all models to make it obvious to user
    else:
        df = df[df['model'] == model_name]

    # calculate and print overall accuracy
    overall_accuracy = len(df[df['model_pred_diff'] >= 0]) / len(df)
    print(f"Overall accuracy for {model_name}: {overall_accuracy}")

    # create figure for density plots
    plt.figure(figsize=(10, 6))
    
    # iterate through linguistic phenomena
    for piece in ["existence", "relations", "actions"]:
        # calculate accuracy for each piece
        piece_accuracy = len(df[(df['model_pred_diff'] >= 0) & (df['linguistic_phenomena'] == piece)]) / len(df[df['linguistic_phenomena'] == piece])
        print(f"{piece} accuracy: {piece_accuracy}")

        # create density plot for model_pred_diff for each piece
        sns.kdeplot(df[df['linguistic_phenomena'] == piece]['model_pred_diff'], linestyle = '--', shade=True, alpha=0.5, label=f'{piece} (Accuracy: {100*piece_accuracy:.1f}%)')
    

    # add OVERALL performance too
    sns.kdeplot(df['model_pred_diff'], shade=False, alpha=0.5, linestyle='-', color = "black", label=f'All three linguistic phenomena (Accuracy: {100*overall_accuracy:.1f}%)')

    # add legend and labels
    # vertical line at 0 (x=0)
    plt.axvline(x=0, color='k', linestyle='--', linewidth=1.5)



    plt.title(f'Model output on entire VALSE subset - {model_name}')
    plt.xlabel('Caption-foil prediction difference')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

graph_performance(df, "CLIP")
graph_performance(df, "LXMERT")
graph_performance(df, "FLAVA")
graph_performance(df, "SigLip")




```
